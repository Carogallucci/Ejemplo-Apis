Importo el archivo json desde https://covidtracking.com/data/api

```python
import requests
import json
import pandas as pd

url='https://api.covidtracking.com/v1/states/daily.json'
respuesta= requests.get(url)
respuesta 
```

```python
respuesta.text
```

Genero el dataframe

```python
texto=respuesta.text
json_data = json.loads(texto)
df = pd.DataFrame.from_dict(json_data)
df
```

```python
df.dtypes
```
